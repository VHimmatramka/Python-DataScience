```python
#importing dataset
df = pd.read_csv('Batting.csv')
```

```python
df1 = df[df['HR']>0].groupby(['birthYear','yearID', 'playerID']).sum()
```

```python
df1.reset_index(inplace=True)
```

## TOTAL NO OF HR IN A YEAR hued as per Player's Birth Year

```python
sns.lmplot(x='yearID',y='HR',data=df1, hue ='birthYear')
```

```python
by_pid = df[df['HR']>0].groupby('playerID').sum()
by_pid.reset_index(inplace=True)
by = by_pid.sort_values(by = 'HR', ascending = False)
by.reset_index(inplace=True)
del by['index']
```

## PLAYERS WITH MOST HR (TOP 5)

```python
sns.barplot(x='playerID',y='HR',data=by.head(5))
```

## Kernel Density Estimation plot of Total HITS scored 


```python
df1['H'].plot.kde()
```

## SPREAD DIAGRAM OF TOP 200 PLAYERS WITH MOST NUMBER OF HITS

```python
df1.sort_values(by='H', ascending = False)
df1.head(200)[['birthYear','H']].iplot(kind='spread')
```
